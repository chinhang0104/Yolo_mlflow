```python
%%writefile {train_src_dir}/train.py
import argparse
import json
import os
import pandas as pd
import mlflow
from ultralytics import YOLO, settings

def select_first_file(path):
    """Selects first file in folder, use under assumption there is only one file in folder
    Args:
        path (str): path to directory or file to choose
    Returns:
        str: full path of selected file
    """
    files = os.listdir(path)
    return os.path.join(path, files[0])

# Start Logging
settings.update({"mlflow": True})

os.makedirs("./outputs", exist_ok=True)

def main():
    """Main function of the script."""

    # input and output arguments
    parser = argparse.ArgumentParser()
    parser.add_argument("--data", type=str, help="path to data")
    parser.add_argument("--epochs", required=False, default=1, type=int)
    parser.add_argument("--model", type=str, help="path to model file")
    parser.add_argument("--training_arg", type=str, help="path to model training arg json")
    args = parser.parse_args()

    # Parse training arg json
    with open(args.training_arg, 'r') as json_file:
        params = json.load(json_file)

    # Load a model
    model = YOLO(args.model)

    # Train the model
    results = model.train(data=args.data, epochs=args.epochs, **params)

    # Register Mlflow 
    # Log the model to MLflow
    mlflow.set_experiment("YOLO_imagenette")
    with mlflow.start_run(run_name="yolo_training"):
        # Log model
        mlflow.pytorch.log_model(model.model, "yolo_model", registered_model_name="YOLO_imagenette",)
    
        # Log metrics
        mlflow.log_metric("fitness", results.fitness)
        mlflow.log_metric("accuracy_top1", results.top1)
        mlflow.log_metric("accuracy_top5", results.top5)
    
        # Log the hyperparameters
        mlflow.log_params(params)

if __name__ == "__main__":
    main()
```
